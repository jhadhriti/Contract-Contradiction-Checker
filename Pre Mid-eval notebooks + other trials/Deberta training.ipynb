```python
from datasets import load_dataset

configs = ['hi', 'bn', 'mr', 'as', 'ta', 'te', 'or', 'ml', 'pa', 'gu', 'kn']
ds_complete = {}
for config in configs:
    ds = load_dataset("Divyanshu/indicxnli", config)
    ds_complete[config] = ds
```

```python
ds_gu = ds_complete['gu']
ds_gu['train']
```

```python
ds_train = {}
ds_test = {}
ds_val = {}

for token in list(ds_complete.keys()):
    ds_train[token] = ds_complete[token]['train']
    ds_test[token] = ds_complete[token]['test']
    ds_val[token] = ds_complete[token]['validation']
```

```python
from torch.utils.data import DataLoader
from sentence_transformers import CrossEncoder
from sentence_transformers.cross_encoder.evaluation import CESoftmaxAccuracyEvaluator, CEF1Evaluator
from sentence_transformers.evaluation import SequentialEvaluator
from sentence_transformers.readers import InputExample
import math
import logging
from datetime import datetime

model = CrossEncoder('cross-encoder/nli-deberta-v3-base')

def conv_2_req(obj):
    return [obj['premise'], obj["hypothesis"]]
scores = model.predict(conv_2_req(ds_val['hi'][10]))

logger = logging.getLogger(__name__)

#Convert scores to labels
label_mapping = ['contradiction', 'entailment', 'neutral']
scores.argmax(), ds_val['hi'][10]["label"]
# labels = [label_mapping[score_max] for score_max in scores.argmax(axis=1)]
```

```python
train_batch_size = 25
num_epochs = 1
```

```python
dev_samples = []
train_samples = []
for item in ds_val["hi"]:
    dev_samples.append(InputExample(texts=[item['premise'], item['hypothesis']], label=item['label']))
    
for item in ds_train["hi"]:
    train_samples.append(InputExample(texts=[item['premise'], item['hypothesis']], label=item['label']))
    
train_batch_size = 16
num_epochs = 4
model_save_path = "output/training_allnli-" + datetime.now().strftime("%Y-%m-%d_%H-%M-%S")

# Define our CrossEncoder model. We use distilroberta-base as basis and setup it up to predict 3 labels
model = CrossEncoder("cross-encoder/nli-deberta-v3-base", num_labels=3)

# We wrap train_samples, which is a list of InputExample, in a pytorch DataLoader
train_dataloader = DataLoader(train_samples, shuffle=True, batch_size=train_batch_size)

softMaxevaluator = CESoftmaxAccuracyEvaluator.from_input_examples(dev_samples, name='softmax_eval')
f1_evaluator = CEF1Evaluator.from_input_examples(dev_samples, name="AllNLI-dev")
evaluator = SequentialEvaluator([softMaxevaluator, f1_evaluator])
```

```python
import torch
from torch import nn

warmup_steps = math.ceil(len(train_dataloader) * num_epochs * 0.1)  # 10% of train data for warm-up
logger.info("Warmup-steps: {}".format(warmup_steps))

torch.cuda.empty_cache()
model.model = nn.DataParallel(model.model, device_ids=[0, 1])
model.model.save_pretrained = model.save_pretrained

# Train the model
model.fit(

    train_dataloader=train_dataloader,
    evaluator=softMaxevaluator,
    epochs=num_epochs,
    evaluation_steps=10000,
    warmup_steps=warmup_steps,
    output_path=model_save_path,
)
```
