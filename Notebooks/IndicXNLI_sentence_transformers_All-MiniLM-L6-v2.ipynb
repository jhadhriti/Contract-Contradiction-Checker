```python
from sentence_transformers import SentenceTransformer

model_id = "sentence-transformers/all-MiniLM-L6-v2"
model = SentenceTransformer(model_id)
```

```python
#from datasets import load_dataset

#dataset_id = "embedding-data/QQP_triplets"
#dataset = load_dataset(dataset_id)
from datasets import load_dataset

configs = ['hi', 'bn', 'mr', 'as', 'ta', 'te', 'or', 'ml', 'pa', 'gu', 'kn']
ds_complete = {}
for config in configs:
    ds = load_dataset("Divyanshu/indicxnli", config)
    ds_complete[config] = ds
```

```python
ds_train = {}
ds_test = {}
ds_val = {}

for token in list(ds_complete.keys()):
    ds_train[token] = ds_complete[token]['train']
    ds_test[token] = ds_complete[token]['test']
    ds_val[token] = ds_complete[token]['validation']
```

```python
from sentence_transformers.readers import InputExample

dev_samples = []
train_samples = []
for item in ds_val["hi"]:
    dev_samples.append(InputExample(texts=[item['premise'], item['hypothesis']], label=item['label']))

for item in ds_train["hi"]:
    train_samples.append(InputExample(texts=[item['premise'], item['hypothesis']], label=item['label']))
```

```python
from torch.utils.data import DataLoader

train_dataloader = DataLoader(train_samples, shuffle=True, batch_size=16)
```

```python
from sentence_transformers import losses

train_loss = losses.SoftmaxLoss(num_labels=3, sentence_embedding_dimension=384,model=model)
```

```python
model.fit(train_objectives=[(train_dataloader, train_loss)], epochs=3)
```
